```python
import numpy as np
import pandas as pd

def calculate_stock_usage(added_stock, stocks_needed_week):
    """
    Calculates the daily stock usage, stock shortages, and stock in storage for a week.

    Args:
        added_stock (numpy.ndarray): A 1D array of length 7 representing the stocks added each day of the week.
        stocks_needed_week (int or float): The total stocks needed for the week.

    Returns:
        tuple: A tuple containing three NumPy arrays:
            - stocks_used (numpy.ndarray): A 1D array of length 7 representing the stocks used each day.
            - stock_shortages (numpy.ndarray): A 1D array of length 7 representing the stock shortages each day.
            - stock_in_storage (numpy.ndarray): A 1D array of length 7 representing the stock in storage each day.
    """
    n_days = len(added_stock)
    stocks_needed = round(stocks_needed_week / n_days)

    stocks_used = np.zeros(n_days)
    stock_shortages = np.zeros(n_days)
    stock_in_storage = np.zeros(n_days)

    for i in range(n_days):
        if stock_in_storage[i] < stocks_needed:
            if added_stock[i] + stock_in_storage[i] >= stocks_needed:
                stock_in_storage[i] = added_stock[i] + stock_in_storage[i] - stocks_needed
                stocks_used[i] = stocks_needed
            else:
                stocks_used[i] = added_stock[i] + stock_in_storage[i]
                stock_shortages[i] = stocks_needed - stocks_used[i]
                if n_days - i - 1 > 0:
                    stocks_needed += round(stock_shortages[i] / (n_days - i - 1))
                else:
                    stocks_needed += round(stock_shortages[i])
        else:
            stock_in_storage[i] = stock_in_storage[i] - stocks_needed + added_stock[i]
            stocks_used[i] = stocks_needed

        stock_in_storage[i:] = stock_in_storage[i]

    return stocks_used, stock_shortages, stock_in_storage

# Iterate over each unique ID in the 'ID' column
result_list = []
for id, group in non_nan_merged_df.groupby('ID'):
    stock_in_storage = 0
    # Split the data into weeks of 7 days
    week_data = [group.iloc[i:i+7] for i in range(0, len(group), 7)]

    # Iterate over the weekly data
    for i, week in enumerate(week_data):
        # Calculate the stocks added each day of the week
        stocks_added_week = week['num_stocks'].values + week['produced_at_plant_9321'].values + week['produced_at_plant_9323'].values
        # Add stock in storage from the previous week to the first day's stock added
        stocks_added_week[0] += stock_in_storage

        # Get the stocks needed for the week
        stocks_needed_week = week['stocks Needed'].values[0]

        # Call the function and get the results
        stocks_used, stock_shortages, stock_in_storage_week = calculate_stock_usage(stocks_added_week, stocks_needed_week)
        
        # Set the stock in storage for the next week as the storage at the end of the current week
        stock_in_storage = stock_in_storage_week[-1]

        # Collect results in a list
        for day, (added, used, shortage, storage) in enumerate(zip(stocks_added_week, stocks_used, stock_shortages, stock_in_storage_week)):
            result_list.append({
                'ID': id,
                'Week': i + 1,
                'Day': day + 1,
                'Stocks Added': added,
                'Stocks Used': used,
                'Stock Shortages': shortage,
                'Stock in Storage': storage
            })

# Convert the results to a DataFrame
results_df = pd.DataFrame(result_list)

stock_num = results_df[results_df['ID'] == 310002952]
stock_num
```
