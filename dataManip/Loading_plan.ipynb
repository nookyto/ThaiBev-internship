```python
import numpy as np
import pandas as pd
from scipy.optimize import minimize

def stock_optimization(total_stocks_needed, added_stock, n_days):
    daily_avg_stock_set = total_stocks_needed / n_days

    def objective(x):
        return np.sum((x - daily_avg_stock_set) ** 2)

    constraints = [{'type': 'eq', 'fun': lambda x: np.sum(x) - total_stocks_needed}]
    bounds = [(0, None) for _ in range(n_days)]
    x0 = np.full(n_days, daily_avg_stock_set)

    result = minimize(objective, x0, method='SLSQP', bounds=bounds, constraints=constraints)
    optimized_stock_set = result.x

    total_stocks_in_system = 0
    stocks_used = np.zeros(n_days)
    stocks_left = np.zeros(n_days)
    stock_shortages = np.zeros(n_days)

    for i in range(n_days):
        total_stocks_in_system += added_stock[i]
        stocks_used[i] = min(optimized_stock_set[i], total_stocks_in_system)
        stocks_left[i] = total_stocks_in_system - stocks_used[i]
        stock_shortages[i] = max(0, optimized_stock_set[i] - stocks_used[i])
        total_stocks_in_system -= stocks_used[i]

    return optimized_stock_set, stocks_used, stocks_left, stock_shortages

# Assuming merged_df is the DataFrame you're starting with
n_days = 7

# Initialize lists to store results
results = []

# Group by ID
grouped = merged_df.groupby('ID')

# Apply optimization for each group
for id_value, group in grouped:
    if len(group) < n_days:
        print(f"Skipping group with ID {id_value} because it has less than {n_days} days of data.")
        continue

    added_stock = group['num_stocks'].values + group['produced_at_plant_9321'].values + group['produced_at_plant_9323'].values 
    total_stocks_needed = group['stocks Needed'].sum()  # Sum total stocks needed across the group
    
    # Ensure that the length of the added_stock array matches n_days
    added_stock = added_stock[:n_days]
    
    optimized_stock_set, stocks_used, stocks_left, stock_shortages = stock_optimization(total_stocks_needed, added_stock, n_days)
    
    results.append(pd.DataFrame({
        'Day': group['Day'].values[:n_days],
        'ID': id_value,
        'optimized_stock_set': optimized_stock_set,
        'stocks_used': stocks_used,
        'stocks_left': stocks_left,
        'stock_shortages': stock_shortages
    }))

# Concatenate results into a single DataFrame
if results:
    results_df = pd.concat(results)
    # Merge results back with the original DataFrame
    merged_df = merged_df.merge(results_df, on=['Day', 'ID'], how='left')

    print(merged_df)

    # Calculate the total stocks used for verification
    total_stocks_used = merged_df['stocks_used'].sum()
    total_stocks_needed_all = merged_df['stocks Needed'].sum()
    print("\nTotal Stocks Needed: ", total_stocks_needed_all)
    print("Total Stocks Used: ", total_stocks_used)
    print("Difference: ", total_stocks_needed_all - total_stocks_used)
else:
    print("No valid groups for optimization.")
```
